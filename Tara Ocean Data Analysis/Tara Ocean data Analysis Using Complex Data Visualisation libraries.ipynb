```python
import pandas as pd

# Load the OTU table
otu_table = pd.read_csv('Tara_OTUtableTax_Full.csv')

# Load the sample metadata
sample_metadata = pd.read_csv('Tara_SampleMeta.csv')

```

```python
otu_table
```

```python
df = otu_table.iloc[:, 6:]
```

```python
mask = otu_table.astype(bool)
```

```python
non_zero_values = otu_table[mask]
```

```python
non_zero_values
```

```python
mask = df.iloc[:,7:].apply(lambda x: sum(x != 0) >= 1, axis=1)

# Create a new DataFrame with the non-zero value and index for rows with multiple non-zero values
df_new = pd.DataFrame({
    'non_zero_index': df[mask].apply(lambda x: list(x[x != 0].index), axis=1),
    'non_zero_value': df[mask].apply(lambda x: list(x[x != 0].values), axis=1)
})
# Display the new DataFrame
print(df_new)
```

```python
df_new.reset_index()
df_new['OTU.rep'] = df_new['non_zero_value'].apply(lambda x: x[0])
df_new['non_zero_value'] = df_new['non_zero_value'].apply(lambda x: x[1:])
df_new['non_zero_index'] = df_new['non_zero_index'].apply(lambda x: x[1:])
```

```python
df3 = pd.merge(otu_table.iloc[:,0:7], df_new, on='OTU.rep')
df3[df3['OTU.rep']=='AB089126.1.1285']
```

```python
exploded = df3.explode(['non_zero_index','non_zero_value'])
```

```python
exploded = exploded.rename(columns={'non_zero_index': 'Sample_ID', 'non_zero_value': 'OTU_value'})
```

```python
combined = pd.merge(exploded, sample_metadata, left_on='Sample_ID',right_on='SampleID', how='left')
```

```python
combined_grp = combined.groupby('Genus').count()
```
